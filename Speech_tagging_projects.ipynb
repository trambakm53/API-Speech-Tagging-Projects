#### The program creates an temporary database and stores user info on this, once the user info is registered the user can  choose functions from the below module.  

*   Named Entity Recognition
*   Language Detection
*   Sentiment Analysis


```python
from johnsnowlabs import nlp
import requests
import json
import emoji

class NLP:

  def __init__(self):

    self.__database = {}
    self.__first_menu()

  def __first_menu(self): ## the functions allows,the user to choose what he wants

    first_input = input("""
    Hi! how would you like to proceed?
    1. Not a member? Register
    2. Already a member? Login
    3. Came by mistake Exit
    """)

    if first_input == '1':
      self.__register()
    elif first_input == '2':
      self.__login()
    else:
      exit()

  def __second_menu(self):

    second_input = input("""
    Hi! how would you like to proceed?
    1. NER
    2. Language Detection
    3. Sentiment Analysis
    4. Logout
    """)

    if second_input == '1':
      self.__ner()
    elif second_input == '2':
      self.__language_detection()
    elif second_input == '3':
      self.__sentiment_analysis()
    else:
      exit()

  def __register(self): ## the functions registers an user
    name = input('enter name')
    email = input('enter email')
    password = input('enter password')

    if email in self.__database:
      print('email already exists')
    else:
      self.__database[email] = [name,password]
      print('registration successful.Now login')
      print(self.__database)
      self.__first_menu()


  def __login(self): ## Once the user is registered, he cna login

    email = input('enter email')
    password = input('enter password')

    if email in self.__database:
      if self.__database[email][1] == password:
        print('login successful')
        self.__second_menu()
      else:
        print('wrong password.Try again')
        self.__login()
    else:
      print('This email is not registered')
      self.__first_menu()

  def __ner(self): ## this function is used for Named-entity-recognition
    para = input("Enter the paragraph")
    #search_term = input("What would you like to search")
    response = nlp.load('ner').predict(para)
    print(response)

  def __sentiment_analysis(self): ## this function is used for sentiment analysis,this identifies whether the text positive,negatiev or neutral
    url = "https://twinword-twinword-bundle-v1.p.rapidapi.com/sentiment_analyze/"
    querystring = {"text":"I've lost count of the number of times I have seen this movie, but it is more than 20. It has to be one of the best movies ever made. It made me take notice Morgan Freeman and Tim Robbins like I had never noticed any actors before."}
    headers = {
	     "X-RapidAPI-Key": "f56ae622fdmsh221087b50806710p125459jsn76859970430e",
	     "X-RapidAPI-Host": "twinword-twinword-bundle-v1.p.rapidapi.com"
     }
    rs = requests.get(url, headers=headers, params=querystring)
    # print(rs.json())
    empty_list = []
    k = rs.json()
    l = k['keywords']
    for m in l:
       #print(m)
       empty_list.append(m['score'])
    index= sorted(list(enumerate(empty_list)), key = lambda x:x[1], reverse = True)[0:3][0]
    #print(emoji.emojize('Python is :grinning_face:'))
    print(l[index]['word'], emoji.emojizef('Python is :grinning_face:'))
    self.__second_menu()

  def __language_detection(self): ## this function is used for language detection, what kind of language is this.
    import requests
    import json
    url = "https://text-analysis12.p.rapidapi.com/language-detection/api/v1.1"
    #payload = { "text": "Hello how are you? Have you heard about SpaceX?" }
    enter_text = input("Enter the text that you want to check")
    payload = { "text": enter_text }
    headers = {
        "content-type": "application/json",
        "X-RapidAPI-Key": "f56ae622fdmsh221087b50806710p125459jsn76859970430e",
        "X-RapidAPI-Host": "text-analysis12.p.rapidapi.com"
    }
    response = requests.post(url, json=payload, headers=headers)
    output = response.json()
    lis = []
    print(output)
    language = output['language_probability']
    print(language)
    for key, values in language.items():
      print(f"The language is: {key} and the confidence is {values}")

obj = NLP()
```
